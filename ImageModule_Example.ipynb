# Image Module â€” Example Usage Notebook
This notebook demonstrates how to use your `Image` class for basic image manipulations. It generates example images so you can run everything out of the box.

**Note:** Some methods in the module have limitations; below we showcase the ones that work as-is.

```python
import sys, os, cv2, numpy as np, matplotlib.pyplot as plt
print('Python', sys.version)
print('CWD:', os.getcwd())
print('OpenCV', cv2.__version__)
```

## 1) Import the module

```python
# The module file is expected to be in the same directory as the notebook
from Image_module import Image

img = Image('test1.png')
print('Loaded image shape (H, W, C):', img.image.shape)
```

## 2) Display the original image

```python
# Display uses matplotlib and may show BGR colors as-is.
img.display('Original (as loaded)')
```

## 3) Blur by averaging

```python
img = Image('test1.png') 
img.blur_by_averaging()
img.display('Averaging blur (5x5 kernel)')
```

## 4) Stronger blur (OpenCV blur)

```python
img = Image('test1.png')  
img.stronger_blur(15)
img.display('Stronger blur (15x15)')
```

## 5) Sharpen

```python
img = Image('test1.png')  
img.sharpen()
img.display('Sharpened')
```

## 6) Invert colors

```python
img = Image('test1.png')  
img.invert_color()
img.display('Inverted Colors')

# Save the result as an example
img.save('inverted_demo.png')
print('Saved to output_images/inverted_demo.png')
```

## 7) Add text overlay

```python
img = Image('test1.png')  
img.add_text('Hello, Image Module!', position=(30, 100), font_size=1, color=(0, 255, 0), thickness=3)
img.display('With Text')
img.save('with_text.png')
print('Saved to output_images/with_text.png')
```

## 8) Add a logo

```python
img = Image('test1.png')  
img.add_logo('logo.png')
img.display('With Logo')
img.save('with_logo.png')
print('Saved to output_images/with_logo.png')
```

## 9) Blend with another image

```python
# Create a second base image variant for blending
variant = cv2.GaussianBlur(cv2.imread('test1.png'), (31, 31), 0)
cv2.imwrite('output_images/second.png', variant)

img = Image('test1.png') 
img.blend_image('second.png', blend_rate1=0.7, blend_rate2=0.5) 
img.display('Blended')
img.save('blended.png')
print('Saved to output_images/blended.png')
```
